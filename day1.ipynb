```julia
function update_dial(x::Int, dir::AbstractString, n::Int, count_moduli::Bool=false)
    if dir == "L"
        y = x - n
        traversed = x:-1:y
    else
        y = x + n
        traversed = x:1:y
    end
    
    r = mod(y, 100)
    
    if count_moduli
        crossings = sum(mod.(traversed[2:end-1], 100) .== 0)
        return (r, crossings)
    else
        return r
    end
end
```

```julia
function match_instructions(instructions::AbstractVector{<:AbstractString})
    regex = r"^([LR])(\d+)"
    matches = match.(regex, instructions)
    output = [(m.captures[1], parse(Int, m.captures[2])) for m in matches]
    return output
end
```

```julia
function get_password(instructions::AbstractVector{<:AbstractString}, start::Int=50)
    directions = match_instructions(instructions)
    update_fns = [x -> update_dial(x, d[1], d[2]) for d in directions]
    positions = accumulate((x, f) -> f(x), update_fns, init = start)
    n_zeros = sum(positions .== 0)
    return n_zeros
end
```
