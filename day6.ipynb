```julia
function do_column(numbers::AbstractArray, op::AbstractString)
    if op == "*"
        return reduce(*, numbers)
    elseif op == "+"
        return reduce(+, numbers)
    elseif op == "/"
        return reduce(/, numbers)
    elseif op == "-"
        return reduce(-, numbers)
    else
        error("Unrecognized cephalopod math op!!")
    end
end
```

```julia
function cephalopod_transpose(numbers::AbstractArray{<:Real})
    powers = 0:size(numbers,1)-1
    power_split = [mod.(div.(numbers, 10^p), 10) for p in powers]
    power_split = filter.(x -> x != 0, power_split)
    numbers = [sum(p .* 10 .^ ((length(p)-1):-1:0)) for p in power_split]
    return numbers
end
    
```

```julia
function do_homework(numbers::AbstractMatrix{<:Real}, ops::AbstractVector)
    return sum([do_column(numbers[:,i], ops[i]) for i in 1:length(ops)])
end
```

```julia
function do_cephalopod_homework(numbers::AbstractMatrix{<:Real}, ops::AbstractVector)
    return sum([do_column(cephalopod_transpose(numbers[:,i]), ops[i]) for i in 1:length(ops)])
end
```
